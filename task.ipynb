# Задание №1

```python
source_type_dict = {
    'organic': ['yandex' , 'google'],
    'ad': ['paid ', 'email']
} 
with open('visit_log.csv') as csv_log:
    log = csv.DictReader(csv_log, delimiter=';')
    with open('visit_log_1.csv', 'w', newline='') as new_csv_log:
        fieldnames = ['timestamp', 'visit_id','url', 'region', 'user_id', 'traffic_source', 'source_type']
        writer_log = csv.DictWriter(new_csv_log, fieldnames=fieldnames, delimiter=';')
        writer_log.writeheader()
        
        for row in log:

            if row['traffic_source'] in source_type_dict['organic']:
                row['source_type']='organic'
                writer_log.writerow(row)
            elif (row['traffic_source'] in source_type_dict['ad']) and (row['region']=='Russia'):
                row['source_type']='ad'
                writer_log.writerow(row)
            elif (row['traffic_source'] in source_type_dict['ad']) and (row['region'] != 'Russia'):
                row['source_type']='other'
                writer_log.writerow(row)
            else:
                row['source_type']=row['traffic_source']
                print(row['source_type'])
                writer_log.writerow(row)
table = pd.read_csv('visit_log_1.csv', sep=';')
table.head()
```

# Задание №3

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], unit='s')
ratings.head()
def life_time(timestamp):
    return timestamp.max() - timestamp.min()
#ratings.groupby('userId').apply(life_time)
user = ratings.groupby('userId').agg({'rating': 'count', 'timestamp': life_time}).reset_index()
user[user['rating']>100]
```
